```python
import re
import numpy as np

def extract_processor(text):
    text = str(text)

    text = re.sub(r'.*Battery', '', text, flags=re.IGNORECASE)

    pattern = r'((Snapdragon|Dimensity|Helio|Exynos|Tensor|Unisoc|MediaTek|Mediatek|Qualcomm|Apple|Bionic|A\d+|T\d+|SC\d+|MTK|MAD|Single Core|Octa Core|Quad Core|Dual Core|[\d]+\s*Core)[\w\s+.-]*?(Processor|Chip))'

    match = re.search(pattern, text, re.IGNORECASE)

    if match:
        return match.group(1).strip()

    return np.nan


df["Processor"] = df["features"].apply(extract_processor)
```

```python
def extract_rear_camera(text):
    text = str(text)
    
    match = re.search(r'(\d+(\.\d+)?MP(?:\s*\+\s*\d+(\.\d+)?MP)*)\s*Rear Camera', text)
    
    if match:
        return match.group(1)
    
    return np.nan

df["Rear_Camera_MP"] = df["features"].apply(extract_rear_camera)

```

```python
def extract_front_camera(text):
    text = str(text)
    
    if "|" in text:
        front_part = text.split("|")[1]
        match = re.search(r'(\d+(\.\d+)?MP(?:\s*\+\s*\d+(\.\d+)?MP)*)', front_part)
        if match:
            return match.group(1)
    
    match = re.search(r'(\d+(\.\d+)?MP(?:\s*\+\s*\d+(\.\d+)?MP)*)\s*Front Camera', text)
    if match:
        return match.group(1)
    
    return np.nan

df["Front_Camera_MP"] = df["features"].apply(extract_front_camera)
```

```python


def extract_warranty_years(text, keyword_type="phone"):
    text = str(text)
    
    if keyword_type == "phone":
        pattern = r'(\d+)\s*(Year|Years|Month|Months).*?(phone|device|handset)'
    else:
        pattern = r'(\d+)\s*(Year|Years|Month|Months).*?(box|accessories|inbox)'
    
    match = re.search(pattern, text, re.IGNORECASE)
    
    if match:
        value = int(match.group(1))
        unit = match.group(2).lower()
        
        if "month" in unit:
            return value / 12
        else:
            return value
    

    if keyword_type == "phone":
        match2 = re.search(r'(\d+)\s*(Year|Years|Month|Months)', text, re.IGNORECASE)
        if match2:
            value = int(match2.group(1))
            unit = match2.group(2).lower()
            
            if "month" in unit:
                return value / 12
            else:
                return value
    
    return np.nan
```

```python
df["Phone_Warranty_Years"] = df["features"].apply(lambda x: extract_warranty_years(x, "phone"))

df["Box_Warranty_Years"] = df["features"].apply(lambda x: extract_warranty_years(x, "box"))
```

```python
def warranty_category(value):
    if pd.isna(value) or value == 0:
        return "No Warranty"
    elif value <= 0.5:
        return "Up to 6 Months"
    elif value <= 1:
        return "1 Year"
    else:
        return "More than 1 Year"

df["Phone_Warranty_Category"] = df["Phone_Warranty_Years"].apply(warranty_category)
df["Box_Warranty_Category"] = df["Box_Warranty_Years"].apply(warranty_category)
```

```python
def extract_battery(text):
    match = re.search(r"(\d{4,5})\s*mAh", str(text), re.IGNORECASE)
    return int(match.group(1)) if match else np.nan

df["Battery_mAh"] = df["features"].apply(extract_battery)
```

```python
def extract_display(text):
    text = str(text)
    
    match = re.search(r'\((\d+(\.\d+)?)\s*inch\)', text, re.IGNORECASE)
    
    if match:
        return float(match.group(1))
    

    match2 = re.search(r'(\d+(\.\d+)?)\s*inch', text, re.IGNORECASE)
    
    if match2:
        return float(match2.group(1))
    
    return np.nan

df["Display_Inches"] = df["features"].apply(extract_display)
```

```python
def convert_to_snake_case(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.replace(r'([a-z])([A-Z])', r'\1_\2', regex=True)
        .str.lower()
        .str.replace(r'[^\w\s]', '', regex=True)
        .str.replace(r'\s+', '_', regex=True)
    )
    return df
```
